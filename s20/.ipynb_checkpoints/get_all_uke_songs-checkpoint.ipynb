```python
songs
```

```python
def song_url_to_json(url):
    url_data = url.split('/')
    artist = url_data[-3]
    song = url_data[-2]
    req = requests.get(url)
    txt = req.text
    lyrics = txt.split('<div id="cont">')[1]
    rgx = '<uke-chord.*?>'
    finds = re.findall(rgx, lyrics)
    chords = []
    for f in finds:
        ch_rgx = 'name=\"\w+'
        chord = re.search(ch_rgx, f).group().split('"')[1]
        chords.append(chord)
    song_dict = {
        'artist': artist,
        'song': song,
        'chords':chords
    }
    j_d = json.dumps(song_dict)
    return j_d
```

```python
song_url =  'https://ukutabs.com/b/billie-eilish/ocean-eyes/'
s = song_url_to_json(song_url)
print(s)
```

```python
f = open("songs.json", "w")
idx = 0
f.write("[\n")
while(idx < len(songs)):
    song = songs[idx]
    try:
        sj = song_url_to_json(song)
        idx += 1
        print("{:3.2f}".format((idx/5084)*100) + "% done " + song, end="\r")
        f.write(sj)
        f.write(',\n')
    except:
        continue
f.write("]")
f.close()
```
