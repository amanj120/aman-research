```python
import requests
import re
import time
import json
```

```python
url = 'https://ukuchords.com/ukulele-chord-diagrams/#viewchooser'
req = requests.get(url)
txt = req.text
rgx = '<uke-chord.*?>'
all_chords = re.findall(rgx,txt)
all_chords
```

```python
print(all_chords[0])
#sub means which version of the chord it is
```

```python
name_r = 'name=\"\S+\"'
fret_r = 'frets=\"\S+\"'
fing_r = 'fingers=\"\S+\"'
c_dict = list()
for c in all_chords:
    name = re.search(name_r, c).group()[6:-1]
    frets = re.search(fret_r, c).group()[7:-1]
    try:
        fingers = re.search(fing_r, c).group()[9:-1].replace('_','0')
        chord = {}
        chord['name'] = name
        chord['frets'] = frets
        chord['fingers'] = fingers
        c_dict.append(chord)
#         print(name, " ", frets, " ", fingers)
    except:
        print(name, ' ', frets, " NO FINGERS FOUND")
        
em6 = {
    "name": "Em6",
    "frets": "0102",
    "fingers": "0102",
}
c_dict.append(em6)
```

```python
json.dumps(c_dict[0])
```

```python
f = open('chords.json', 'w')
f.write('[\n')
for item in c_dict:
    f.write(json.dumps(item) + ',\n')
f.write(']')
f.close()
```
