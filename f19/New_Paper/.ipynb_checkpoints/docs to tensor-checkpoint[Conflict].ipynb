```python
import nltk
from nltk.corpus import stopwords
# nltk.download('stopwords')
import os
from os import listdir
from os.path import isfile, join
import itertools
import numpy as np
from numpy import log as ln
```

```python
abspath = 'C:/Users/amanj/Google Drive/Current/Research/New_Paper/parsed_docs/'
stops = set(stopwords.words('english'))
punctuation = {',', '(', ')', '-', "'", '"',"", '/', '\\', '~', '`', ';', '*', '!', '?', '@','#', '^', '&', '+', '_', '\n', '"', "'"}
stops = stops.union(punctuation)
random_words_im_too_lazy_to_fix_the_correct_way = { 'mart=c3=adn', 'win=e2=80=a6'
, 'win=e2=80=a6', 'yes=e2=80=a6', 'championships=e2=80=a6','=3d', 
'=e2=85=93', '=f0=9f=8c=8e', '=f0=9f=90=98', '=f0=9f=a4=af'}
stops = stops.union(random_words_im_too_lazy_to_fix_the_correct_way)
allWords = set();
# stops
```

```python
#print(os.path.abspath(abspath))
files = [f for f in listdir(abspath) if isfile(join(abspath, f))]
#len(files) #this is supposed to be 1043
```

```python
absolute_counts_of_all_words = {} #stores word: (index, count)
wordIdx = 0
numWords = 0;
#going to go through all of the words and see how many of them there are
```

```python
for i in range(len(files)):
    f = open(abspath + files[i], 'r')
    parsed = f.read().lstrip().rstrip()
    sentences = parsed.split('.')[:-1] #get all of the sentences
    for s in sentences:
        words_in_sentence = s.split(' ')
        for w in words_in_sentence:
            word = w.lower()
            if word not in stops:
                numWords += 1
                if word in absolute_counts_of_all_words:
                    absolute_counts_of_all_words[word] = (absolute_counts_of_all_words[word][0], absolute_counts_of_all_words[word][1] + 1)
                else:
                    absolute_counts_of_all_words[word] = (wordIdx, 1)
                    wordIdx += 1;
```

```python
tensor = {}
# (i,j,k):count
fileIndex = 0;
mi = 100000
ma = -100000
for fname in files:
    k = fileIndex;
    fileIndex = fileIndex + 1;
    f = open(abspath + fname, 'r')
    file = f.read().lstrip().rstrip()
    sentences = file.split('.')[:-1] # get all sentences
    numPairs = 0;
    pairsInFile = {} #{(word1, word2) : count} dictionary
    for s in sentences:
        s_words = s.split(' ')
        l = len(s_words)
        for i in range(l - 1):
            w1 = s_words[i]
            if w1 in absolute_counts_of_all_words:
                for j in range(i + 1, l):
                    w2 = s_words[j]
                    if w2 in absolute_counts_of_all_words:
                        if(w1 != w2):
                            numPairs += 1
                            key = (w1,w2)
                            if key in pairsInFile:
                                pairsInFile[key] += 1
                            else:
                                pairsInFile[key] = 1
    sumNum = 0
    for key in pairsInFile.keys():
        v = pairsInFile[key]
        numerator = v/numPairs
        p1 = absolute_counts_of_all_words[key[0]][1]/numWords
        p2 = absolute_counts_of_all_words[key[1]][1]/numWords
        idx1 = absolute_counts_of_all_words[key[0]][0]
        idx2 = absolute_counts_of_all_words[key[1]][0]
        denominator = p1*p2
        t_key = (idx1, idx2, k)
        tensor[t_key] = ln(numerator/denominator)
        mi = min(mi, tensor[t_key])
        ma = max(ma, tensor[t_key])
        sumNum += numerator
print(sumNum)
print(mi)
print(ma)
         
#     #print(file)
#     for s in sentences:
#         s_words = s.split(' ')
#         filtered = list(filter(lambda x: x in d.keys(), s_words))
#         #print(filtered)
#         for pair in itertools.product(filtered, filtered):
#             w1 = pair[0]
#             w2 = pair[1]
#             if(w1 != w2):
#                 index = (d[w1][0], d[w2][0],k)
#                 denom = d[w1][1] * d[w2][1]
#                 if index in tensor:
#                     tensor[index] = tensor[index] + (1/denom)
#                 else:
#                     tensor[index] = (1/denom)
# len(tensor.keys())
```
